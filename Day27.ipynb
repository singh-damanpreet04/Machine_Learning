```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.DataFrame({
    'No_of_Rooms': [3, 4, 5, 2],
    'No_of_Grocery_Shops': [2, 0, 6, 10],
    'Price_Lakhs': [60, 130, 170, 90]
})

print("Original Data:")
print(data)
```

```python
# Step 2: Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 3: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Step 4: Create a DataFrame of reduced data
reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])

print("\nReduced Data using PCA (2 components):")
print(reduced_df)

# Optional: Explained Variance Ratio
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)
```
