```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
house_sizes = np.array([500, 750, 1000, 1250, 1500, 1750, 2000]).reshape(-1, 1)
house_prices = np.array([150, 200, 250, 300, 350, 400, 450])
```

```python
X_train, X_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

```python
plt.scatter(house_sizes, house_prices, color="blue", label="Actual Data")
plt.plot(house_sizes, model.predict(house_sizes), color="red", label="Regression Line")
plt.title("House Price Prediction")
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.show()
```

```python
price_to_predict = np.array([[3000]])
predicted_price = model.predict(price_to_predict)
print(f"Predicted price for a house of size sq ft: ${predicted_price[0]:.2f}")
```
