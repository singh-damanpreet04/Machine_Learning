```python
import seaborn as sns
```

```python
df= sns.load_dataset('iris')
```

```python
df
```

```python
df['species'].unique()
```

```python
df.head()
```

```python
#independent and dependent features
x=df.iloc[ : , :-1]
y=iris.target
```

```python
x,y
```

```python
# train test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#post pruning
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(x_train,y_train)
```

```python
#construct tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)  #range of gini 0-0.5
```

```python
# doing hyperparameter tuning and pruning as unnecessary nodes are there
treemodel = DecisionTreeClassifier(max_depth = 2)
```

```python
treemodel.fit(x_train,y_train)
```

```python
#construct tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

in green gini the output will bw versonica and in blue it will be versicolor

```python
#prediction
y_pred=treemodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
print(score)
```

```python
print(classification_report(y_test,y_pred))
```
